```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris
from scipy.cluster.hierarchy import linkage, fcluster, dendrogram
```

```python
iris = load_iris()
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df['sepal length (cm)'] = (df['sepal length (cm)'] - df['sepal length (cm)'].mean()) 
df['sepal length (cm)'].std()
```

```python
df['petal length (cm)'] = (df['petal length (cm)'] - df['petal length (cm)'].mean()) 
df['petal length (cm)'].std()
```

```python
Z = linkage(df[['sepal length (cm)', 'petal length (cm)']], 'ward')
```

```python
labels = fcluster(Z, 3, criterion='maxclust')
```

```python
df['cluster'] = labels
```

```python
plt.figure(figsize=(25, 10))
dendrogram(Z)
plt.title("Iris Dendrogram")
plt.xlabel("Sample Index")
plt.ylabel("Distance")
```

```python
sns.scatterplot(x=df['sepal length (cm)'], y=df['petal length (cm)'], hue=df['cluster'],
palette='Dark2')
plt.title("Iris Clustered Data")
plt.xlabel("Standardized Sepal Length (cm)")
plt.ylabel("Standardized Petal Length (cm)")
plt.show()
```
